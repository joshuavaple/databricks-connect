```python
import mlflow


mlflow.set_tracking_uri("databricks://joshuale-common") # this profile already stores the remote workspace URL

# 2) set the experiment you want to log to
mlflow.set_experiment("/Shared/your-test-experiment") # this will create the experiment if it doesn't exist

# 3) start a run and log stuff
with mlflow.start_run() as run:
    print(run.info.run_name)
    mlflow.log_param("p", 42)
    mlflow.log_param("model_type", "demo")
    
    for i in range(5):
        mlflow.log_metric("metric_val", i * 2)

print("done")
```
