```python
from databricks.sdk import WorkspaceClient


catalog_name = "workspace"
schema_name = "bronze"
vol_name = "demo"

w = WorkspaceClient(profile = "joshuale-common")
fs = w.dbutils.fs
```

## 1. List Files in a Volume

```python
root_dir = f"/Volumes/{catalog_name}/{schema_name}/{vol_name}"
fs.ls(root_dir)
```

## 2. Create a File and Write to Vol

```python
file_path = f"{root_dir}/hello.txt"
file_data = "Hello, Databricks!"
fs = w.dbutils.fs

fs.put(
  file      = file_path,
  contents  = file_data,
  overwrite = True
)

print(fs.head(file_path))
```

```python
fs.ls(root_dir)
```

## 3. Secret Scope
- First, create a scope: https://docs.databricks.com/aws/en/security/secrets/?language=Databricks%C2%A0workspace%C2%A0UI#create-a-secret-scope

```python
# add a secret to a scope
w.secrets.put_secret("demo-scope","sample-key",string_value ="sample-value")
```

```python
# list secret keys in all scopes
for secret_scope in w.dbutils.secrets.listScopes():
    for secret_metadata in w.dbutils.secrets.list(secret_scope.name):
        print(f'found {secret_metadata.key} secret in {secret_scope.name} scope')
```
